## SQL

### 1. Create Table

```python
CREATE TABLE students (
    id INT PRIMARY KEY,        -- 唯一标识
    name VARCHAR(50),          -- 字符串，最大长度50
    age INT,                   -- 整数
    score DECIMAL(5, 2)        -- 小数，如 98.50
);
```

```python
SELECT * FROM students
SELECT name, score, AVG(score) FROM students
SELECT DISTINCT gender FROM students

SELECT * FROM students WHERE score > 80 AND age < 25
SELECT * FROM students WHERE name LIKE '张%'
SELECT * FROM students ORDER BY score DESC

INNER JOIN classes ON students.class_id = classes.id;

GROUP BY gender
ORDER BY score DESC 
HAVING AVG(score) > 80; -- 这里的过滤是针对“平均分”这个聚合结果的

LIMIT 5
LIMIT 10 OFFSET 10（跳过前 10 条，从第 11 条开始取 10 条）


WHERE age != 18
WHERE id IN (1, 3, 5)
WHERE score BETWEEN 60 AND 100
WHERE email IS NULL

```

```python
SELECT gender, COUNT(*) 
FROM students 
GROUP BY gender; -- 按性别分组统计人数

SELECT gender, course, COUNT(*) AS cnt
FROM students
GROUP BY gender, course;
```
