```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Data Read

```python
df = pd.read_csv("sample.csv")
```

```python
df.head()
```

## 2. Visualization

### 2.1 Bar Chart

```python
# Single Variable
#df["feature_1"].head(10).plot(kind="bar")

# Vertical
#df[["feature_1", "feature_2", "feature_3"]].head(10).plot(kind="bar")

#Horizontal
#df[["feature_1", "feature_2", "feature_3"]].head(10).plot(kind="barh")

#Stack
df[["feature_1", "feature_2", "feature_3"]].head(10).plot(kind="bar", stacked=True, figsize=(10, 5))

plt.title("Bar Chart: feature_1–3 (first 10)")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.show()
```

### 2.2 Histogram

```python
#single variable
#df["feature_2"].plot(kind="hist", bins=30)

#multiple variables
df[["feature_5", "feature_6", "feature_7"]].plot(kind="hist",bins=30,alpha=0.6)
plt.title("Histogram: feature_5, feature_6, feature_7")
plt.show()
```

#### (2) Multiple Variable

### 2.3 Pie Chart

```python
df["feature_3"].value_counts(bins=5).plot(kind="pie",autopct="%1.1f%%")

plt.title("Pie Chart: feature_3 (binned)")
plt.ylabel("")
plt.show()
```

### 2.4 Scatter Plot

#### (1)Single Variable

```python
#df_1 = df.head(100)
df_1 = df.iloc[1:200]
plt.scatter(df_1["feature_1"], df_1["feature_2"])

plt.xlabel("feature_1")
plt.ylabel("feature_2")
plt.title("Scatter Plot: feature_1 vs feature_2")
plt.show()

```

#### (2)Multiple Variable

##### i. Pair Compare (1 vs 2 & 1 vs 3)

```python
df_100 = df.head(100)
df_1 = df.iloc[1:200]
plt.figure(figsize=(8,5))

plt.scatter(df["feature_1"], df["feature_2"], label="feature_2")
plt.scatter(df["feature_1"], df["feature_3"], label="feature_3")
plt.scatter(df["feature_1"], df["feature_4"], label="feature_4")

plt.xlabel("feature_1")
plt.ylabel("Values")
plt.title("Scatter Plot: feature_1 vs multiple features")
plt.legend()
plt.show()
```

##### ii. Pair Compare (1 vs 2 & 3 vs 4)

```python
df_100 = df.head(100)
df_1 = df.iloc[1:200]

plt.figure(figsize=(8,5))

plt.scatter(df["feature_1"], df["feature_2"], label="f1 vs f2")
plt.scatter(df["feature_3"], df["feature_4"], label="f3 vs f4")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot: Multiple Variable Pairs")
plt.legend()
plt.show()
```

### 2.5 Line Plot

#### (1) Single Plot

```python
df["feature_4"].head(100).plot()
plt.title("Line Plot: feature_4 (first 100)")
plt.show()
```

#### (2) Multiple Plots

```python
df[["feature_1", "feature_2", "feature_3"]].plot(subplots=True, figsize=(6,6))
plt.show()
```

#### (3) Multiple Variable

##### i. no change to the variable name

```python
df[["feature_1", "feature_2", "feature_3", "feature_4"]].head(100).plot(figsize=(8,5))
plt.title("Multiple Line Plot: feature_1–4")
plt.show()
```

##### ii. change to the variable name

```python
plt.figure(figsize=(8,5))

plt.plot(df["feature_1"].head(100), label="feature_1")
plt.plot(df["feature_2"].head(100), label="feature_2")
plt.plot(df["feature_3"].head(100), label="feature_3")
plt.plot(df["feature_4"].head(100), label="feature_4")

plt.legend()
plt.title("Multiple Line Plot (First 100 Rows)")
plt.show()
```

### 2.6 Box Plot

```python
df[["feature_1", "feature_2", "feature_3"]].plot(kind="box")
plt.title("Box Plot: feature_1–3")
plt.show()
```

## 2.7 Correlation Heatmap

```python
corr = df[["feature_1", "feature_2", "feature_3", "feature_4"]].corr()

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```
