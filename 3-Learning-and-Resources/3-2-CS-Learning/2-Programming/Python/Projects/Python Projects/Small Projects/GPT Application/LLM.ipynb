```python
import openai
import sys
from rich.console import Console

# 初始化终端美化工具
console = Console()

# 设置 OpenAI API Key（请替换成你的 API Key）
client = openai.OpenAI(api_key="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")

# 选择可用模型（你可以修改这里使用 gpt-4o-mini 或 gpt-3.5-turbo）
MODEL_NAME = "gpt-4o"  # 可选 "gpt-4o-mini", "gpt-3.5-turbo"

def ask_chatgpt(prompt):
    """调用 OpenAI API 获取 AI 响应"""
    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "system", "content": "你是一个智能写作助手。"},
                  {"role": "user", "content": prompt}],
        temperature=0.7,
    )
    return response.choices[0].message.content

def save_to_markdown(content, filename="ai_generated_text.md"):
    """将 AI 生成的内容保存到 Markdown 文件"""
    with open(filename, "w", encoding="utf-8") as f:
        f.write(content)
    console.print(f"[bold green]内容已保存至 {filename}[/bold green]")

def main():
    """终端交互主函数"""
    console.print(f"[bold green]欢迎使用 AI 写作 Agent！当前使用模型：{MODEL_NAME}[/bold green]")
    console.print("输入你的写作主题，或输入 'exit' 退出程序")

    while True:
        try:
            user_input = input("> ")  # 用户输入
            if user_input.lower() in ["exit", "quit", "q"]:
                console.print("[bold red]退出 AI 写作 Agent[/bold red]")
                break

            console.print("[bold cyan]AI 正在生成内容...[/bold cyan]")
            result = ask_chatgpt(user_input)
            console.print(f"\n[bold yellow]AI 生成结果：[/bold yellow]\n{result}\n")

            # 选择是否保存
            save_choice = input("是否保存到 Markdown 文件？(y/n): ").strip().lower()
            if save_choice == "y":
                save_to_markdown(result)

        except KeyboardInterrupt:
            console.print("\n[bold red]用户终止程序[/bold red]")
            break

if __name__ == "__main__":
    main()
```
