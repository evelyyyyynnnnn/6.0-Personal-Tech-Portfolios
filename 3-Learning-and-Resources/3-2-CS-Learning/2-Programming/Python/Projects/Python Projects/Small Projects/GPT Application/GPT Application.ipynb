````python
import openai
import requests
import json
import re
import os
import random
from docx import Document
import shutil 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# 设置 OpenAI API Key 和 Google API Key
OPENAI_API_KEY = "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"
GOOGLE_API_KEY = "AIzaSyDdtlF6mDgiWUhEE4i_yn8jvjDkbl-GE-M"
GOOGLE_CSE_ID = "92ecc769947d44a95"
SEARCH_ENGINE_URL = "https://www.googleapis.com/customsearch/v1"
SERPAPI_KEY = "c23a6f9445cdeec3d4ba15e305e98328bcb3fffc8eec68faf07eb8c65783bf5e"
MODEL_NAME = "gpt-3.5-turbo-1106"

client = openai.OpenAI(api_key=OPENAI_API_KEY)


# ----------------------------------------
# Markdown Format
# ----------------------------------------




# ----------------------------------------
# Input Regulization
# ----------------------------------------

def translate_to_english(text):
    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[
            {"role": "system", "content": "You are a translator that translates Chinese text into English."},
            {"role": "user", "content": text}
        ],
        temperature=0.7,
    )
    return response.choices[0].message.content.strip()  # Use dot notation



# ----------------------------------------
# Visualization
# ----------------------------------------

def ask_for_table_plot(topic, table_or_figure_name, type_, save_path):
    os.makedirs(save_path, exist_ok=True)  # Ensure directory exists

    if type_ == "table":
        prompt = f"""
        Generate a well-formatted table with meaningful data for '{table_or_figure_name}' related to the research topic '{topic}'.
        The table should be informative and contain structured data relevant to the topic.
        Return the structured data in JSON format without additional explanation.
        """
        response = ask_chatgpt(prompt)

        try:
            table_data = json.loads(response)
        except json.JSONDecodeError:
            return {"name": table_or_figure_name, "filename": None, "content": f"Failed to parse table data for {table_or_figure_name}"}

        df = pd.DataFrame(table_data)
        table_filename = os.path.join(save_path, f"{table_or_figure_name.replace(' ', '_')}.csv")
        df.to_csv(table_filename, index=False, encoding="utf-8")
        table_content_str = df.to_string(index=False)

        return {
            "name": table_or_figure_name,
            "filename": table_filename,
            "content": table_content_str
        }

    elif type_ == "figure":
        prompt = f"""
        Generate a description and structured data representation for the figure '{table_or_figure_name}' related to the research topic '{topic}'.
        The figure should be informative and present a meaningful visualization relevant to the topic.
        """
        response = ask_chatgpt(prompt)

        try:
            figure_data = json.loads(response)
        except json.JSONDecodeError:
            return {"name": table_or_figure_name, "filename": None, "content": f"Failed to parse figure data for {table_or_figure_name}"}

        df = pd.DataFrame(figure_data)
        plt.figure(figsize=(8, 6))
        df.plot(kind="bar")
        plt.title(table_or_figure_name)
        plt.xlabel("Category")
        plt.ylabel("Values")
        plt.grid(True)
        
        figure_filename = os.path.join(save_path, f"{table_or_figure_name.replace(' ', '_')}.png")
        plt.savefig(figure_filename)
        plt.close()
        figure_content_str = df.to_string(index=False)

        return {
            "name": table_or_figure_name,
            "filename": figure_filename,
            "content": figure_content_str
        }

    else:
        return {"name": table_or_figure_name, "filename": None, "content": "Invalid type specified. Must be 'table' or 'figure'."}



# ----------------------------------------
# Search
# ----------------------------------------

def search_google(query, num_results=10):
    """
    通过 Google Custom Search API 获取搜索结果
    """
    params = {
        "key": GOOGLE_API_KEY,
        "cx": GOOGLE_CSE_ID,
        "q": query,
        "num": num_results
    }

    response = requests.get(SEARCH_ENGINE_URL, params=params)
    
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Google API 请求失败，状态码: {response.status_code}")
        return None


def fetch_full_content(url):
    """
    访问网页并提取正文内容
    """
    headers = {"User-Agent": "Mozilla/5.0"}
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            paragraphs = [p.get_text() for p in soup.find_all("p") if len(p.get_text()) > 50]
            return "\n".join(paragraphs[:5]) if paragraphs else "无法提取网页正文内容"
        else:
            return f"请求失败，状态码: {response.status_code}"
    except Exception as e:
        return f"请求错误: {e}"


def get_combined_research(query, num_results=10):
    """
    获取 Google 搜索的前 num_results 个结果，并提取它们的正文内容，合并成一个文本块。
    """
    search_results = search_google(query, num_results)
    research = ""

    if search_results:
        for item in search_results.get("items", []):
            url = item.get("link", "")
            title = item.get("title", "无标题")
            
            full_content = fetch_full_content(url)  # 爬取网页正文
            research += f"\n标题: {title}\n{full_content}\n\n{'='*80}\n"

    return research  # 返回所有爬取的内容


# ----------------------------------------
# Reference
# ----------------------------------------

def fetch_scholar_papers(query, num_results=25):
    url = "https://serpapi.com/search.json"
    params = {
        "engine": "google_scholar",
        "q": query,
        "api_key": SERPAPI_KEY,
        "num": num_results
    }
    response = requests.get(url, params=params)
    return response.json().get("organic_results", [])

    
def format_to_apa(paper, index):
    authors_list = paper.get("publication_info", {}).get("authors", [])
    formatted_authors = []
    for author in authors_list:
        name_parts = author["name"].split()
        if len(name_parts) > 1:
            formatted_authors.append(f"{name_parts[-1]}-{''.join([p[0] for p in name_parts[:-1]])}")
        else:
            formatted_authors.append(name_parts[0])
    authors = ", ".join(formatted_authors)
    
    year_match = re.search(r'\b(\d{4})\b', paper.get("publication_info", {}).get("summary", ""))
    year = year_match.group(1) if year_match else "n.d."
    
    title = paper.get("title", "Unknown Title")
    title = re.sub(r": (\w)", lambda m: f": {m.group(1).upper()}", title)
    
    summary_parts = paper.get("publication_info", {}).get("summary", "").split(" - ")
    journal = summary_parts[1] if len(summary_parts) > 1 else "Unknown Journal"
    
    doi = paper.get("link", "")
    
    return f"{authors}. ({year}). {title}. {journal}. DOI: {doi}", authors, year, index

def generate_in_text_citations(references):
    citations = []
    for ref in references:
        authors = ref[1].split(", ")
        year = ref[2]
        index = ref[3]
        if len(authors) == 1:
            citation = f"({authors[0]}, {year})[{index}]"
        elif len(authors) == 2:
            citation = f"({authors[0]} et {authors[1]}, {year})[{index}]"
        else:
            citation = f"({authors[0]} et al, {year})[{index}]"
        citations.append(citation)
    return citations

def insert_in_text_citations(text, citations):
    sentences = re.split(r'(\.|!|\?)', text)
    num_citations = min(len(sentences) // 3, len(citations))
    random_indices = random.sample(range(len(sentences)), num_citations)
    for i in sorted(random_indices):
        if not re.match(r'^\s*#+\s', sentences[i]) and citations:  # 确保不是标题
            sentences[i] = sentences[i].strip() + " " + citations.pop(0)
    return "".join(sentences)



# ----------------------------------------
# Main Body
# ----------------------------------------

def ask_chatgpt(prompt):
    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "system", "content": "你是一个智能联网搜索助手。"},
                  {"role": "user", "content": prompt}],
        temperature=0.7,
    )
    return response.choices[0].message.content


def generate_paper():
# ----------------------------------------
# Folder Location
# ----------------------------------------

     
    subfolder_name = input("请输入论文文件夹名称（例如：音乐训练 SCI论文）: ")
    save_path = os.path.join("/Users/evelyndu/Dropbox/缓冲资料库/dx/未交", subfolder_name)
    os.makedirs(save_path, exist_ok=True)
    
    # 模板文件路径
    template_path = "/Users/evelyndu/Dropbox/缓冲资料库/dx/Template/template.docx"
    # 目标路径（复制到论文文件夹中）
    template_copy_path = os.path.join(save_path, "template.docx")
    # 复制模板文件
    shutil.copy(template_path, template_copy_path)
    print(f"模板文件已复制至: {template_copy_path}")

    # ✅ 创建 table_figure 子文件夹
    table_figure_path = os.path.join(save_path, "table_figure")
    os.makedirs(table_figure_path, exist_ok=True)
    print(f"Tables and Figures will be saved in: {table_figure_path}")

    # 创建 "选题.docx" 文件
    topic_doc_path = os.path.join(save_path, "选题.docx")
    
    # 让用户输入选题内容
    topic_content = input("请输入选题内容: ")
    
    # 写入选题内容
    doc = Document()
    doc.add_paragraph(topic_content)
    doc.save(topic_doc_path)
    
    print(f"选题文件已保存至: {topic_doc_path}")
    

    
    topic_CN = input("Enter the research topic in Chinese: ")
    topic_with_quotes = f'"{topic_CN}"'
    topic = translate_to_english(topic_with_quotes)

    # ✅ Call get_combined_research() to fetch real-time research before using it
    research = get_combined_research(topic, num_results=10)

    papers = fetch_scholar_papers(topic)
    references = [format_to_apa(paper, i+1) for i, paper in enumerate(papers[:25])]
    in_text_citations = generate_in_text_citations(references)
    

    prompt = f"""
    Generate a detailed academic outline (at least 3 levels) for the paper titled: 
    '{topic}'.
    The outline should include:
    1. Abstract (300-400 words)
    2. Key Words (3-4)
    3. Introduction (400-500 words)
    4. Literature Review (800-1000 words, at least 3 subsections)
    5. Methodology (detailed structure including empirical analysis)
    6. Results (including at least 5 tables, 3 images, statistical analysis)
    7. Discussion (500-1000 words, at least 3 aspects)
    8. Conclusion (max 500 words)
    9. References (APA format, 25 citations since 2021)
    for each section should have 3 levels following the numerical order like 3.1, 3.1.1
    The latest real-time research is as follows:{research},please refer to them.
"""
    outline = ask_chatgpt(prompt)
    print(outline)
    # ✅ 确保章节顺序正确
    sections = [
        "Abstract", "Key Words", "Introduction", "Literature Review", 
        "Methodology", "Results", "Discussion", "Conclusion"
    ]
    
    # 设定每个部分的基础字数要求
    base_sections = {
        "Abstract": 300,
        "Key Words": 50,
        "Introduction": 550,
        "Literature Review": 1200,
        "Methodology": 1200,
        "Results": 1200,
        "Discussion": 100,
        "Conclusion": 600
    }
    
    # 让字数在 ±200 范围内浮动
    sections = {section: random.randint(words - 200, words + 200) for section, words in base_sections.items()}
    
    paper_content = {}
    material = {}

    for section, words in sections.items():
        if section not in ["Results", "Discussion"]:  # Skip "Results" and "Discussion"
            prompt_text = (
                f"Write a detailed academic section on '{section}' for the topic: {topic}.\n\n"
                f"Word Count Requirement**: {words} words (±200 words allowed).\n\n"
                f"All titles should follow the Markdown structure from the outline:\n\n"
                f"{outline}\n\n"
                f"Formatting Instructions:\n"
                f"- Do NOT include the section title at the beginning, it will be added later.\n"
                f"- Do NOT use section numbering (e.g., '4.1', '4.1.1'). Instead:\n"
                f"  - Level-2 subtitles (e.g., '4.1') should start with '## '.\n"
                f"  - Level-3 subtitles (e.g., '4.1.1') should start with '### '.\n"
                f"The latest real-time research is as follows:{research}, please refer to them.\n"
            )
    
            response = ask_chatgpt(prompt_text)
            paper_content[section] = response

        if section in ["Results", "Discussion"]:
            prompt_text += (
                f"- Include at least **4 tables** and **3 figures** with reasonable simulated data (for the statistical results' table (like correlation), you don't need to find real data, just generate reasonable data).\n"
                f"- The tables and figures should provide **quantitative support** related to '{topic}' with reasonable titles.\n"
                f"- Return only the names of the tables and figures, formatted strictly as a JSON block (without explanation):\n```json\n{{'tables': [...], 'figures': [...]}}\n```"
            )
            
            data= ask_chatgpt(prompt_text)
            match = re.search(r'```json\n(.*?)\n```', data, re.DOTALL)
            if match:
                json_str = match.group(1)
                try:
                    material[section] = json.loads(json_str)
                    print(f"Generated {section} Tables and Figures: {material[section]}")
                except json.JSONDecodeError:
                    print(f"Failed to parse JSON response: {json_str}")
                    material[section] = {"tables": [], "figures": []}
            else:
                print(f"Failed to extract JSON response: {data}")
                material[section] = {"tables": [], "figures": []}
            
            updated_tables = []
            updated_figures = []
            
            for table in material[section]["tables"]:
                table_content = ask_for_table_plot(topic, table, "table", save_path)
                updated_tables.append(table_content)
        
            for figure in material[section]["figures"]:
                figure_content = ask_for_table_plot(topic, figure, "figure", save_path)
                updated_figures.append(figure_content)
            
            material[section]["tables"] = updated_tables
            material[section]["figures"] = updated_figures
            print("material:",material)

            final_prompt = (
                f"Write a refined and detailed section on the Results and Discussion Part of '{section}' for '{topic}', you just need to add their name follow the order like Table 1, Figure 1 at the center of some lines before  commenting on them to interprete and I will copy the content of table of figure later by myself to .\n\n"
                
                f"### Tables\n"
                f"Below are the detailed tables that summarize key aspects of the study:\n\n"
                + "\n\n".join([f"**{table['name']}**\n{table['content']}" for table in material[section]['tables']]) + "\n\n"
                f"Each table contains structured data relevant to the study. Provide an analytical discussion about the patterns and insights they reveal.\n\n"
            
                f"### Figures\n"
                f"The following figures have been generated and saved as PNG files:\n\n"
                + "\n\n".join([f"**{figure['name']}**\n{figure['content']}" for figure in material[section]['figures']]) + "\n\n"
                f"Refer to these figures and analyze how they contribute to the research discussion."
            )


            response = ask_chatgpt(final_prompt)
            paper_content[section] = response


# ----------------------------------------
# Essay Lib
# ----------------------------------------
   
    # Markdown 文件路径
    md_file_path = os.path.join(save_path, f"{topic_CN}.md")

    # 生成 Markdown 论文
    with open(md_file_path, "w", encoding="utf-8") as file:
        for section, content in paper_content.items():
            updated_content = insert_in_text_citations(content, in_text_citations.copy())
            heading_level = 1 if section in ["Abstract", "Key Words", "Introduction", "Literature Review", 
                                             "Methodology", "Results", "Discussion", "Conclusion"] else 2
            file.write(f"{'#' * heading_level} {section}\n\n{updated_content}\n\n")
        file.write("# References\n\n")
        for ref in references:
            file.write(f"[{ref[3]}] {ref[0]}\n\n")
    
    # 生成 Word 文档并从 Markdown 读取内容
    
    doc = Document(template_copy_path)
    with open(md_file_path, "r", encoding="utf-8") as file:
        content = file.read().split("\n")  # 读取整个 Markdown 并按行分割
        for line in content:
            if line.startswith("# "):
                doc.add_heading(line.strip("# ").strip(), level=1)
            elif line.startswith("## "):
                doc.add_heading(line.strip("## ").strip(), level=2)
            elif line.startswith("### "):
                doc.add_heading(line.strip("### ").strip(), level=3)
            elif line.startswith("#### "):
                doc.add_heading(line.strip("#### ").strip(), level=4)
            else:
                doc.add_paragraph(line.strip())
    
    doc.save(template_copy_path)
    print(f"Academic paper generated and saved as '{md_file_path}' and '{template_copy_path}'.")

    

# ----------------------------------------
# Execution
# ----------------------------------------

if __name__ == "__main__":
    generate_paper()
````
