# Data Operation

```python
data=[5,4,3,2,1]
total=sum(data)
mean=sum(data)/len(data)
maximum=max(data)
sorted_data=sorted(data)
print("sum", total)
print("mean", mean)
print("maximum", maximum)
print("sort",sorted_data)
```

```python
import pandas as pd
data={'id':[1,2,3,4,5],'spb':[950,1200,800,1500,1100]}
test=pd.DataFrame(data)
test=test[test['spb']>1000]
test
```

# 2. Visualization

## 1. Scatter

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,6]
y=[5,4,3,2,3]
plt.scatter(x,y)
plt.title("Scatter of x and y")
plt.show()
```

## 2. Line

```python
plt.plot(x,y)
plt.title("line of x and y")
plt.show()
```

## 3. Bar

```python
plt.bar(x,y)
plt.title("bar of x and y")
plt.show()
```

## 4.Hist

```python
plt.hist(y)
plt.title("hist of y")
plt.show()


y=[5,4,3,2,3]
```

## 5.Pie

```python
plt.pie(x)
plt.title("pie of x")
plt.show()
```

## 6.Step

```python
plt.step(x,y)
plt.title("steps of x and y")
plt.show()
```
